```python
#Q3
# There are three types of errors in arithmetic error class
# ZeroDivisionError
# OverflowError
# FloatingPointError


# ZeroDivisionError
try:
    10/0
except ZeroDivisionError as e:
    print(e)
    
```

```python
# OverflowError
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except OverflowError as e:
    print(e)
```

```python
#Q4
#when an index or a key is not found for a list or dictionary then 
#lookup error class is used to handle these type of exceptions.

#Index error

x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(e)
```

```python
#Key error
info = {'name': 'Tanishq Singh',
                'age': 23,
                'language': 'Python'}
try:
    print(info['city'])
except LookupError as e:
    print(e,e.__class__)
```
